```python
import requests


def get_area_id(area_name):
    # API 配置
    url = "https://weather01.market.alicloudapi.com/day15"
    headers = {
        "Authorization": "APPCODE f7620a3f53bd42fa9929a4d8972f2a2b"
    }  # 请替换成你的 AppCode

    # 请求参数
    params = {"area": area_name}
    try:
        # 发送 GET 请求
        response = requests.get(
            url,
            params=params,
            headers=headers,
            verify=False,  # 对应 curl 的 -k 参数，忽略 SSL 证书验证
        )

        # 检查响应状态
        response.raise_for_status()

        # 打印响应头
        print("响应头:")
        for key, value in response.headers.items():
            print(f"{key}: {value}")

        print("\n响应内容:")
        print(response.text)

        # 如果需要返回 JSON 数据
        return response.json()

    except requests.exceptions.RequestException as e:
        print(f"请求错误: {e}")
        return None
```

```python
# 调用函数
if __name__ == "__main__":
    result = get_area_id("广州")
    if result:
        print("\n解析后的 JSON 数据:")
        print(result)
```
